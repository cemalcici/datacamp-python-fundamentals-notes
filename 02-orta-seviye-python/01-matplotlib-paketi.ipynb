## Matplotib ile Basit Grafikler

Veri analizi, veri bilimi için çok önemlidir.

- Veriyi keşfetmeyi sağlar.
- Analiz yapmayı sağlar.

```python
import matplotlib.pyplot as plt

yil = [1950, 1970, 1990, 2010]
nufus = [2.519, 3.692, 5.263, 6.972]
plt.plot(yil, nufus) # Yıl ve Nüfus çizgi grafiği oluşturuldu.
plt.show()
```

```python
plt.scatter(yil, nufus) # Yıl ve Nüfus saçılım grafiği oluşturuldu.
plt.show()
```

```python
plt.scatter(yil, nufus) # Yıl ve Nüfus saçılım grafiği oluşturuldu.
plt.plot(yil, nufus) # Yıl ve Nüfus çizgi grafiği oluşturuldu.
plt.show()
```

## Histogram

Seri dağılımı hakkında bilgi verir.

```python
degerler = [0, 0.6, 1.4, 1.6, 2.2, 2.5, 2.6, 3.2, 3.5, 3.9, 4.2, 6]
plt.hist(degerler, bins=3) # 3 bölmeli histogram oluşturuldu
plt.show()
```

## Düzenleme

Veri görselleştirmede birçok yöntem mevcuttur.

- Farklı grafik türleri
- Grafik türlerine göre farklı düzenleme araçları.

Burada hangi grafik türünü ve araçları seçmemiz gerektiğine iş problemine ve veriye bakarak karar verebiliriz.

```python
import matplotlib.pyplot as plt

yil = [1950, 1970, 1990, 2010]
nufus = [2.519, 3.692, 5.263, 6.972]
plt.plot(yil, nufus) # Yıl ve Nüfus çizgi grafiği oluşturuldu.
plt.show()
```

```python
plt.plot(yil, nufus) # Yıl ve Nüfus çizgi grafiği oluşturuldu.
plt.xlabel('Yıl') 
plt.ylabel('Nüfus')
plt.show()
```

```python
plt.plot(yil, nufus) # Yıl ve Nüfus çizgi grafiği oluşturuldu.
plt.xlabel('Yıl') 
plt.ylabel('Nüfus')
plt.title('Yıllara Göre Dünya Popülasyon Göstergesi')
plt.show()
```

```python
plt.plot(yil, nufus) # Yıl ve Nüfus çizgi grafiği oluşturuldu.
plt.xlabel('Yıl') 
plt.ylabel('Nüfus')
plt.title('Yıllara Göre Dünya Popülasyon Göstergesi')
plt.yticks([0, 2, 4, 6, 8]) # Y koordinatının gösterge listesi
plt.show()
```

```python
plt.plot(yil, nufus) # Yıl ve Nüfus çizgi grafiği oluşturuldu.
plt.xlabel('Yıl') 
plt.ylabel('Nüfus')
plt.title('Yıllara Göre Dünya Popülasyon Göstergesi')
plt.yticks([0, 2, 4, 6, 8],
           ['0', '2M', '4M', '6M', '8M']) # Y koordinatının etiketleriyle birlikte gösterge listesi
plt.show()
```

```python
# Yeni Veriler Eklendi
yil = [1800, 1850, 1900] + yil 
nufus = [1.0, 1.262, 1.650] + nufus

plt.plot(yil, nufus) # Yıl ve Nüfus çizgi grafiği oluşturuldu.
plt.xlabel('Yıl') 
plt.ylabel('Nüfus')
plt.title('Yıllara Göre Dünya Popülasyon Göstergesi')
plt.yticks([0, 2, 4, 6, 8],
           ['0', '2M', '4M', '6M', '8M']) # Y koordinatının etiketleriyle birlikte gösterge listesi
plt.show()
```
